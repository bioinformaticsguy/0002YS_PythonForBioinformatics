## File Iteration

```
with open(filename) as file:
    for line in file:
        do something with the line
```

````python
with open("seqdump.txt") as file:
    for line in file:
        print(line)
````

## Dictionary Iteration

```
for key in dictionary.keys():
    do something with key
```

````python
dictionary = {"one": 1, "two": 2, 'three': 3}

for key in dictionary.keys():
    print(key)
````
