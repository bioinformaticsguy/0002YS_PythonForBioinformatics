```python
def validate_base_sequence(base_sequence, RNAFlag = False):
    """ Return True if the sting base_sequence contains only upper or lower case
    T (or U in case of RNAFlag), A, G and C characters, otherwise False """
    DNABases = {"T","A","G", "C"}
    RNABases = {"U","A","G", "C"}
    return set(base_sequence) <= (RNABases if RNAFlag else DNABases)
```

```python
print(validate_base_sequence("ATGAG", True))
```
